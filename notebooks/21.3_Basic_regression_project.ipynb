
## Algerian Forest Fires Dataset
#### About Dataset
The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class)
The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

#### The dataset's attributes are detailed below:
1. Temporal AttributesDay: Day of the month (1 to 31).
- Month: Month of the year (June to September).
- Year: Year of the observation (e.g., 2012).
2. Meteorological ObservationsThese attributes track the daily weather conditions at noon (when temperature is typically at its daily maximum).
- Temperature: Maximum noon temperature in Celsius degrees (22 to 42).
- RH: Relative Humidity in percentage (21 to 90).
- Ws: Wind speed in km/h (6 to 29).
- Rain: Total daily rainfall in mm (0 to 16.8).
3. Fire Weather Index (FWI) ComponentsThese six standard FWI sub-indices measure the moisture content of the soil and different types of forest fuels:
- FFMC: Fine Fuel Moisture Code (28.6 to 92.5).
- DMC: Duff Moisture Code (1.1 to 65.9).
- DC: Drought Code (7 to 220.4).
- ISI: Initial Spread Index (0 to 18.5).
- BUI: Buildup Index (1.1 to 68).
- FWI: Fire Weather Index (0 to 31.1).
4. Target VariableClasses: A binary classification indicating either "Fire" or "not Fire".

## Data Cleaning

```python
## missing value
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1:"Bejaia Region Dataset'

2: "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python

dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.info()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
## fix spaces in colulmns names
df.columns=df.columns.str.strip()
df.columns
```

#### Change the required columns to integer  data type

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

#### Change the other columns to float datatype

```python
objects=[feature for feature in df.columns if df[feature].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i]=df[i].astype(float)
```
